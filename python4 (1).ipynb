```python
df.columns=df.columns.str.strip().str.lower()
```

```python
# columns where 0 is not a valid medical vallue

invalid_zero_cols = [
    'glucose',
    'bloodpressure',
    'skinthickness',
    'insulin',
    'bmi'
]

# replace 0 with nan
df[invalid_zero_cols] = df[invalid_zero_cols].replace(0, np.nan)
```

```python
# impute missing values

for col in invalid_zero_cols:
    df[col].fillna(df[col].median(),inplace=True)# impute missing values
```

```python
# feature target separation

x = df.drop('outcome',axis=1)

# target
y = df['outcome']
```

```python
x.columns
```

```python
y.name
```

```python
# train and test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
   x,y,test_size=0.2,random_state=42,stratify=y
  )
```

```python
x_train.shape,x_test.shape
```

```python
y.value_counts(normalize=True)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# here using logistic regression classifier because it is a
# binary classification problem, strong baseline for  medical data

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train_scaled,y_train)

# here now the model train, uses scaled features, can implement safely
```

```python
# predictions

y_pred = log_reg.predict(x_test_scaled)
```

```python
y_pred[:10]
```

```python
log_reg.score(x_test_scaled,y_test)

# accuracy of test
```

```python
# accuracy of train
log_reg.score(x_train_scaled,y_train)
```

```python
# linear regression probability prediction
y_prob = log_reg.predict_proba(x_test_scaled)
```

```python
y_prob[:20]
```

```python
y_prob_diabetic = y_prob[:,1]
y_prob_diabetic[:10]
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
# [TN,FP]
# [FN,TP]

# TN - true negative - actually non diabetic
# TP - true positive - actually non diabetic but predicted  diabetic
# FN - false negative - actually diabetic but predicted non diabetic
# PP - false positive - actually diabetic
```

```python
# classification report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

# here we will see about precision, recall, f1 score and support
# precision: how correct the model was when it said it said 0 or 1
# recall: how many real 0 and 1 it found
# files:balance score of precision and recall
# support : how many actual 0 and 1 patients

# example for weighted avg
# classes with more samples influence the final score more than the smaller classes
```
