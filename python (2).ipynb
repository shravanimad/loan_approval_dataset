```python
# finding the avg in first case

avg_score = df['G3'].mean()
avg_score

#finding the above avg

above_avg_students= df[df['G3'] > avg_score]
above_avg_students[['G3']]

#count the no of go above avg

above = (df['G3']> avg_score).sum()

print("Above avg :", above)
```

```python
# find how many students passed the final exam

passed = (df['G3'] >= 10).sum()

print("Passed :", passed)
```

```python
# maplotlib
# matplotlib tool is helpful for visualization
# histogram

import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['G3'], bins = 10)
plt.xlabel('G3')
plt.ylabel('Frequency')
plt.title('Histogram of G3')
plt.show()
```

```python
df.columns
```

```python
#barchart
#avg final mark score

avg_score_by_gender = df.groupby('sex')['G3'].mean()
avg_score_by_gender

plt.figure()
avg_score_by_gender.plot( kind = 'bar')
plt.xlabel('Gender')
plt.ylabel('Average score')
plt.title('Average score by Gender')
plt.show()
```

```python
# scatter plot

plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.xlabel('studytime')
plt.ylabel('G3')
plt.title('Scatter plot of G1 vs G3')
plt.show()
```

```python
# box plot

plt.figure()
plt.boxplot(df['G3'])
plt.ylabel('G3')
plt.title('spread of final score')
plt.show()
```

```python
# line plot
avg_marks=df[['G1','G2','G3']].mean()
plt.figure()
plt.plot(avg_marks.index,avg_marks.values,marker='o')
plt.xlabel("subject")
plt.ylabel("average marks")
plt.title("average marks by subject")
plt.show()
```

```python
# with bar chart we will find the number of students
# who failed abd passed the final exam

pass_ = df['G3' ]>= 10

plt.figure()
plt.bar(['pass','fail'],[pass_.sum(), (~pass_).sum()])
plt.ylabel('Number of students')
plt.title('Number of student who pass or fail ')
plt.show()
```

```python
# histogram -
plt.figure()
plt.hist(df['absences'], bins = 10)
plt.xlabel('Absences')
plt.ylabel('Frequency')
plt.title('Histogram of Absences')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram 1 - black bars with white edges
plt.figure()
plt.hist(df['absences'], bins=10, color='skyblue', edgecolor='brown')
plt.xlabel("Absences")
plt.ylabel("Number of Students")
plt.title("Absences Distribution")
plt.show()
```
